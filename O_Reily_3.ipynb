```python
np.random.seed(0)

def funk(vals):
    outp = np.empty(len(vals))
    for i in range(len(vals)):
        outp[i] = 1.0 / vals[i]
    return outp
```

```python
big_array = np.random.randint(1, 100, size=100000)
%timeit funk(big_array)
%timeit print(1.0 / big_array)
```
